```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import Request,urlopen
import requests
import urllib
import re
```

```python
base = "https://www.chicagomag.com"
param ="/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago"

hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(base+param,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
# soup = soup.prettify()
```

```python
article_body = soup.find("div", class_="article-body")
rank = soup.select('div.article-body>div.sammy>.sammyRank')
meta = soup.select('div.article-body>div.sammy>.sammyListing>a')
title = soup.select('div.article-body>div.sammy>.sammyListing>a>b')

link_parsed = []

for i in range(len(meta)):
    if len(re.split((".com"),meta[i]["href"])) > 1:
        link_parsed.append(base + re.split((".com"),meta[i]["href"])[1])
    else:
        link_parsed.append(base + re.split((".com"),meta[i]["href"])[0])

metadata = {
    'name' : [re.split(("\n|\r\n"),meta[i].get_text())[0] for i in range(len(meta))],
    'dish' : [re.split(("\n|\r\n"),meta[i].get_text())[1] for i in range(len(meta))],
    'rank' : [rank[i].get_text() for i in range(len(rank))],
    'link' : link_parsed,
    }
df = pd.DataFrame(metadata)
df
```
