```python
def d_make_sentence(dic) :
    d_ret = []
    if not "@" in dic : return "no Dic"
    d_top = dic["@"]

    w1 = d_word_choice(d_top)
    w2 = d_word_choice(d_top[w1])

    d_ret.append(w1)
    d_ret.append(w2)

    while True :
        try:
            w3 = d_word_choice(dic[w1][w2])
            print( w3 )
            d_ret.append(w3)
        except Exception as e :
            print("error : ", e)
            break
        
        if w3 == "." : break
        w1, w2 = w2, w3
    d_ret = "".join(d_ret)

def d_word_choice(sel) :
    d_keys = sel.keys()
    return random.choice(list(d_keys))

d_make_sentence(dic)
```
