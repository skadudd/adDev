```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as no
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species'] = iris.target
```

```python
df
```

```python
plt.figure(figsize=(5,2))
sns.boxplot(x='sepal length (cm)',y='species',data=df,orient='h')
plt.figure(figsize=(5,2))
sns.boxplot(x='sepal width (cm)',y='species',data=df,orient='h')
plt.figure(figsize=(5,2))
sns.boxplot(x='petal length (cm)',y='species',data=df,orient='h')
plt.figure(figsize=(5,2))
sns.boxplot(x='petal width (cm)',y='species',data=df,orient='h')
plt.show();
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(df,hue='species');
plt.show();
```

```python
sns.pairplot(data=df,vars=['petal length (cm)','petal width (cm)'],hue='species',height=4)
plt.show();
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df,hue='species',palette='Set2')
plt.show();
```

```python
iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:,2:],iris.target)

y_pred_tr = iris_tree.predict(iris.data[:,2:])
accuracy_score(iris.target,y_pred_tr)
```
